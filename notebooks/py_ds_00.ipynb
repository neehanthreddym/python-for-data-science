# Data Filtering and Cleaning

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
num_df = DataFrame(np.arange(0,90,3).reshape(10,3), index = ['row1', 'row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8', 'row9', 'row10'], columns=['col1', 'col2', 'col3'])

num_df
```

### Simple indexing

```python
num_df.iloc[0,1]
```

```python
num_df.iloc[0,1] = 10
num_df
```

### Fancy indexing

```python
num_df.iloc[[2,3,4], [1,2]]
```

### Boolean indexing
Using Comparision operators and Masking

```python
mask = num_df > 28
mask
```

```python
num_df[mask]
```

```python
num_df[num_df<28] = 0
num_df
```

```python
num_df.iloc[2:5, 1:3]
```
