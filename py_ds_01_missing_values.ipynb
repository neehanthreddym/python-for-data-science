```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

### **Filling missing values using fillna(), replace(), interpolate()**

```python
data = {'names': ['john', 'micheal', 'steve', 'sarah', 'emily', 'richard', 'julie'],
        'age': [28, 34, 32, 26, 26, 35, 30],
        'gender': ['male', 'male', 'male', 'female', 'female', 'male', 'female'],
        'rank': [1, 3, 5, 7, 2, 4, 6]}

ranking_df = DataFrame(data)
ranking_df.iloc[2:5, 1] = np.nan
ranking_df.iloc[3:6, 3] = np.nan
ranking_df.iloc[3, :] = np.nan

ranking_df
```

```python
ranking_df.isnull()
```

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
ranking_df.fillna(0)
```

```python
ranking_df.fillna(method='pad')
```

```python
ranking_df.bfill()
```

```python
ranking_df.interpolate(method='linear')
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all')
```

```python
ranking_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```

```python
ranking_df.replace(np.nan, 0)
```
